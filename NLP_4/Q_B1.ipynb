# Î’. GRAPH-BASED DEPENDENCY PARSER

The given code implements a graph-based dependency parser based on the work of [Kiperwasser and Goldberg (2016) ](https://transacl.org/ojs/index.php/tacl/article/viewFile/885/198). The model consists of a biderectional LSTM encoder. Two MLPs are utilized for two different goals. The first one detects the type of the dependency while the second one measures the confidence (score) of the dependency. Several changes inside the model as well as BERT encoder are leveraged to improve the perfomance of the parser.

We execute main.py with --n_lstm_layers 1 as argument since encoder should have only one layer of BiLSTM. Also, the arguments --do_eval --model_dir "path_to_trained_model" are required to examine the performance on the test dataset.  

The experiments require 5 training epochs. The las and uas evaluation metrics are illustrating the ability of the model to make correct predictions.   
las: 92.07
uas: 93.55

The uas score is significantly better than the best scores of the original transition-based dependency parser (Q1, Q4 uas: 89.09). In conclustion, graph based model is preferred.
